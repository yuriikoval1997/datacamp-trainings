# Binomial distribution

## Bernoulli trials

```python
def perform_bernoulli_trials(n: int, p: float):
    """Perform n Bernoulli trials with success probability p
    and return number of successes."""
    n_success = 0

    # Perform trials
    for i in range(n):
        # Choose random number between zero and one: random_number
        random_number = np.random.random()

        # If less than p, it's a success so add one to n_success
        if random_number < p:
            n_success += 1
    return n_success
```

```python
probability = .05
size = 1000
n_defaults = np.empty(size)
for i in range(size):
    n_defaults[i] = perform_bernoulli_trials(100, probability)
```

```python
plt.hist(n_defaults, edgecolor='black')
plt.xlabel('number of defaults out of 100 loans')
plt.ylabel('probability')
plt.show()
```

```python
def my_ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n + 1) / n
    return x, y
```

```python
x, y = my_ecdf(n_defaults)

plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('number of defaults')
plt.ylabel('probability')
plt.show()
```

## Binomial

```python
samples = np.random.binomial(100, .05, size=1000)
```

```python
x, y = my_ecdf(samples)
plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('number of defaults')
plt.ylabel('probability')
plt.show()
```

```python
bins = np.arange(0, np.max(samples) + 1.5) - 0.5
plt.hist(samples, density=True, bins=bins, edgecolor='black')
plt.xlabel('number of defaults')
plt.ylabel('probability')
plt.show()
```
