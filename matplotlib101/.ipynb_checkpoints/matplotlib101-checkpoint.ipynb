```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# matplotlib.pyplot
`plt.subplots` - returns 2 objects:
* `fig: Figure` - is a container that holds everything you see on the page 
* `ax: Axes` - contains data

```python
fig, ax = plt.subplots()
plt.show()
```

```python
seattle_weather = pd.read_csv('datasets/seattle_weather.csv')
seattle_weather.iloc[:, :8]
```

```python
seattle_weather_by_month = seattle_weather[['MONTH', 'MLY-TAVG-NORMAL']].groupby(['MONTH']).mean()
seattle_weather_by_month.columns = ['avg of monthly avg']
seattle_weather_by_month
```

```python
austing_weather = pd.read_csv('datasets/austin_weather.csv')
austing_weather_by_month = austing_weather[['MONTH', 'MLY-TAVG-NORMAL']].groupby(['MONTH']).mean()
austing_weather_by_month.columns = ['avg of monthly avg']
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather_by_month, marker='o', linestyle='--', color='orange')
ax.plot(austing_weather_by_month, marker='v', linestyle=None) # Instead of None you can use empty string
ax.set_xlabel('Time (month)')
ax.set_ylabel('Average temperature of monthly average')
ax.set_title("I don't have enough imagination to create title name")
plt.show()
```

### Small multiples

```python
fig, ax = plt.subplots(3, 2, sharey=True) # sharey - means all subplots will have the same y-scale
ax[0, 0].plot(seattle_weather_by_month, color='b')
ax[1, 1].plot(austing_weather_by_month, color='orange')
plt.show()
```

```python
# (2, 1) - special case
fig, ax = plt.subplots(2, 1)
ax[0].plot([1, 2, 3], color='b')
ax[1].plot([4, 5, 6], color='r')
plt.show()
```

```python
# (1, 2) - special case
fig, ax = plt.subplots(1, 2)
ax[0].plot([1, 2, 3], color='b')
ax[1].plot([4, 5, 6], color='r')
plt.show()
```
