```python
cars = pd.read_csv('datasets/cars.csv', index_col=0)
cars
```

```python
# If you specify header=None, pandas won't consider first row as a header
cars = pd.read_csv('datasets/cars.csv', header=None)
cars
```

```python
cars = pd.read_csv('datasets/cars.csv', header=None, names=['Country', 'drives_right', 'cars_per_cap'])
cars
```

```python
# let's mark values from the first row as NaN, because we 'names' argument to pass columns names
cars = pd.read_csv('datasets/cars.csv', 
                   header=None, 
                   names=['country', 'drives_right', 'cars_per_cap'],
                   na_values=['country', 'drives_right', 'cars_per_cap'])
cars
```

```python
# na_values can also accept a dictionary.
# There might be a case when csv file contains comments. Then you would like to specify the comment start symbol.
# If you know that there is a column containing dates, you should use `parse_dates` option.
cars = pd.read_csv('datasets/cars.csv', header=None, names=['country', 'drives_right', 'cars_per_cap'],
                  na_values={
                      'country': ['country'], 
                      'drives_right': ['drives_right'], 
                      'cars_per_cap': ['cars_per_cap']
                  }, comment='#', parse_dates=True)
cars
```
